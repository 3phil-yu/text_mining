**1.2.分词处理**

使用jieba库将其进行分词处理

```python
comment = df_comment["评论"]
commentlist = []
for i in comment:
    commentlist.append(i)
print(commentlist[:100])
```

```python
content_all_segment = []
for content_each in commentlist:
    try:
        content_each_segment = jieba.lcut(content_each) #jieba.lcut(): 精确分词，将句子最精确的分开，返回词后的列表
        if len(content_each_segment) > 1 and content_each_segment != '\r\n': #过滤掉分词失败的项
            content_all_segment.append(content_each_segment)
        else:
            print(content_each_segment)
    except:
        print(content_each, ' fail to cut')
        continue
```

**1.3.停用词处理**

读取停用词表，去除停用词

```python
stopwords = pd.read_csv("data_stopwords.txt",index_col=False,quoting=3,names=['stopword'], encoding='utf-8')
#stopwords = stopwords.append(pd.read_csv("data/data_stopwords_Weibo.txt",index_col=False,quoting=3,names=['stopword'], encoding='utf-8'))
#index_col=False: 使dataframe不用第一列作为索引。
#names: 设置列名
#quoting: 若不设置quoting，默认会去除英文双引号，只留下英文双引号内的内容，设置quoting = 3，会如实读取内容。
stopwords
```

```python
#编写去停用词函数
def drop_stopwords(contents,stopwords):
    contents_clean = []
    all_words = []
    for line in contents: #遍历contents，对每一条评论进行去停用词
        line_clean = []
        for word in line:
            if word in stopwords or not '\u4e00' <= word <= '\u9fff': #过滤非中文
                continue
            line_clean.append(word) #将本条评论的所有非停用词写入line_clean
            all_words.append(str(word)) #将本条评论的所有非停用词写入all_words，最终得出全部评论的非停用词，方便之后计算词频
        contents_clean.append(line_clean) #将本条评论的全部非停用词写入contents_clean
    return contents_clean,all_words #返回contents_clean与all_words

stopwords = stopwords.stopword.values.tolist()
contents_all_segment_clean, contents_all_words = drop_stopwords(content_all_segment,stopwords)
```

```python
filelist = []
for a in df_contents_all_segment_clean["contents_all_segment_clean"]:
    filelist.append(a)
filelist = str(filelist)
```

```python

filelist2 = []
char_list = []
for a_string in filelist:
    char_list.append(a_string)
    #print(char_list)

    list1 = []
    list1.append(char_list[0])
    list2 = ['']

    #记录要删除的索引
    del1 = []
    i = 0
    while (i<len(char_list)):
        i = i+1
        #这里是对后面没有词汇的时候对列表1和列表2判断一次重复
        if i == len(char_list):
            if list1 == list2:
                m = len(list2)
                for x in range(i-m,i):
                    del1.append(x)
        else:
            
            if char_list[i] == list1[0] and list2==['']:
                #print('词汇和list1相同，list2为空，将词加入list2')
                list2[0]=char_list[i] #这里初始化用append会让lisr2初始化为['','**']
            elif char_list[i] != list1[0] and list2==['']:
                #print('词汇和1不一样，2为空，将词加入1')
                list1.append(char_list[i])

            #触发判断
            elif char_list[i] != list1[0] and list2 !=['']:
                if list1 == list2 and len(list2)>=2:
                    #print('词和1不一样，2不为空，判断1和2重复')
                    m = len(list2)
                    #删除列表2里的内容，列表1原本的内容不用再去判断重复了
                    for x in range(i-m,i):
                        del1.append(x)
                    list1= ['']
                    list2 = ['']
                    list1[0]=char_list[i]
                else:
                    #print('词和1不一样，2不为空，判断1和2不重复')
                    list2.append(char_list[i])

            #触发判断
            elif char_list[i] == list1[0] and list2 != ['']:
                if list1 == list2:
                   # print('词和1相同，2不为空，判断1和2重复')
                    m = len(list2)
                    #删除列表2里的内容，列表1须要再去和后面的词汇继续判断重复
                    for x in range(i-m,i):
                        del1.append(x)

                    list2 = ['']
                    list2[0]=char_list[i]
                else:
                    #print('词和1相同，2不为空，判断1和2不重复')
                    #逻辑对书本上进行了修改，书上是清空列表1和2，就是保留如今列表1和2内容不作删除，这里只保留1，列表2内容还须要作对比
                    list1 = list2
                    list2 = ['']
                    list2[0]=char_list[i]


    a = sorted(del1) #从数字更大的索引删起，这样就不用考虑元素删除后索引的变化问题
    t = len(a) - 1
    while(t>=0):
        del char_list[a[t]]
        t = t-1
    str1 = ''.join(char_list)
    str2 = str1.strip()
    filelist2.append(str2)

print(filelist2)
```

**3.1.词袋模型函数使用举例**

将提取的词汇转化为模型要求的格式：list of list形式，做出映射词袋

```python
texts=[['美国','中国','关系','美国'],['美国','疫情','疫情'],['中国', '疫情'], ['疫情']]
dictionary = corpora.Dictionary(texts) 
for key, value in dictionary.iteritems():
    print(key, value)
#dictionary中，自动将全部词汇生成语料库，并将其设置key值，不同词对应不同的key，例如中国的key为0，关系的key为1，
#美国的key为2，疫情的key为3。
```

```python
corpus = [dictionary.doc2bow(text) for text in texts] #bow: Bag-of-words
corpus
#corpus中，将每条文本的分词结果转化为以下形式:在第一条文本中，key为0的词出现了1次，key为1的词出现了1次，key为2的词出现了2次
```

**3.2.建立词袋模型**

```python
dictionary = corpora.Dictionary(contents_all_segment_clean) #将之前分完词的语料库转为对应字典词袋格式
corpus = [dictionary.doc2bow(contents_each_segment_clean) for contents_each_segment_clean in contents_all_segment_clean]
#把每一条评论都转为语料库词袋模型
```

```python
corpus[0][:5]
#第一条评论的前五个词袋模型
```

**3.3.建立LDA主题模型**

建立LDA模型，设置参数

```python
# corpus：每条博文的语料  id2word：映射字典  num_topics：将评论聚为5类
model_lda = gensim.models.ldamodel.LdaModel(corpus=corpus, id2word=dictionary, num_topics=5)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer
n_features = 1000

tf_vectorizer = CountVectorizer(strip_accents = 'unicode',
                                max_features=n_features,
                                stop_words='english',
                                max_df = 5,
                                min_df =1)#CountVectorizer统计词频
tf = tf_vectorizer.fit_transform(df_comment["评论"].head(40))
tf
```

```python
from sklearn.decomposition import LatentDirichletAllocation
for i in range(5):
    n_topics = i
    lda = LatentDirichletAllocation(#n_topics=n_topics, max_iter=50,
                                #learning_method='online',
                                #learning_offset=50.,
                                )#random_state=0
    lda.fit(tf)
    print(i,lda.perplexity(tf))
```
